```python
# Import modules
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV
census_csv = "Energy Census and Economic Data US 2010-2014.csv"
census_df = pd.read_csv(census_csv)

atlantic_csv = "atlantic.csv"
atlantic_df = pd.read_csv(atlantic_csv)
```

```python
# Census dataset
census_df.head()
```

```python
# Filter desired states

# Create mask for desired states
df_filter = census_df['StateCodes'].apply(lambda state: state in ['MN','NH','VT','MA','RI','CT','NY','NJ','PA','MD','OH','VA','WV','NC','SC','GA','FL','US'])

# Apply filter to census_df
census_df = census_df[df_filter]
census_df.head(20)
```

```python
# Create function for filtering columns
base_strings = ['TotalP','CoalP','GeoP','HydroP','ElecPrice','NatGasPrice','LPGPrice']

def select_columns(base_string,columns):
    for year in range(2010,2015):
        columns.append(base_string + str(year))
    return columns

columns2 = []
for base_string in base_strings:
    select_columns(base_string,columns2)

print(columns2)
```

```python
# Columns for df
final_columns = ['StateCodes', 'Region', 'Division'] + columns2
df = census_df[final_columns]
df
```

```python
# Export cleaned dataframe
df.to_csv('census_data.csv')
```
